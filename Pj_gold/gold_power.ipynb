### 1.2. Проверим, что эффективность обогащения рассчитана правильно. 

Наименование признаков: [этап].[тип_параметра].[название_параметра]. Пример: rougher.input.feed_ag

Блок [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Блок [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

```python
# создадим функцию для расчета эффективности обогащения

def recovery(rougher_output, rougher_input, rougher_tail):
    recovery = (rougher_output * (rougher_input - rougher_tail)) \
    / (rougher_input * (rougher_output - rougher_tail)) * 100
    return recovery
```

```python
# вычислим эффективность обогащения на обучающей выборке для признака rougher.output.recovery

recovery = recovery(gold_industry_train['rougher.output.concentrate_au'],
                    gold_industry_train['rougher.input.feed_au'],
                    gold_industry_train['rougher.output.tail_au'])
```

```python
print(mean_absolute_error(gold_industry_train['rougher.output.recovery'], recovery))
```

Эффективность обогащения рассчитана правильно, MAE между нашими расчётами и значением признака составила всего 9.83758577644259e-15.

### 1.3. Проанализируем признаки, недоступные в тестовой выборке.

```python
# Находим разность множеств названий колонок
different_columns = set(gold_industry_train.columns) - set(gold_industry_test.columns)

# Выводим названия колонок, которые не совпадают
print('Количество отсутствующих признаков в тестовой выборке:', len(list(different_columns)))
print('')
print(list(different_columns))
```

```python
def count_words_with_output(lst):
    input = 0
    output = 0
    state = 0
    calculation = 0
    for item in lst:
        words = item.split('.')
        for word in words:
            if 'input' in word:
                input += 1
            if 'output' in word:
                output += 1
            if 'state' in word:
                state += 1
            if 'calculation' in word:
                calculation += 1       
    return print('input:', input,'output:', output, 'state:', state, 'calculation:', calculation)
```

```python
distrib_au = []
distrib_ag = []
distrib_pb = []
    
def distrib(val):
    for item in val:
        words = item.split('.')
        for word in words:
            if 'feed_au' in word or 'concentrate_au' in word:
                distrib_au.append(item)
            if 'feed_ag' in word or 'concentrate_ag' in word:
                distrib_ag.append(item)
            if 'feed_pb' in word or 'concentrate_pb' in word:
                distrib_pb.append(item)   
```

## 3. Модель


### 3.1. Напишем функцию для вычисления итоговой sMAPE.

```python
def sMAPE(target, predict):
    sMAPE = np.mean((abs(target - predict) / ((abs(target) + abs(predict)) / 2)) * 100)
    return sMAPE
    
```

```python
def sMAPE_tot(rougher, final):
    sMAPE_tot = 0.25 * sMAPE(rougher) + 0.75 * sMAPE(final)
    return sMAPE_tot
```
